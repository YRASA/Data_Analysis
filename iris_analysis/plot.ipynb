```python
import pandas
iris = pandas.read_csv("iris.csv")
```

```python
iris.sample(10)
```

```python
iris.describe()
```

```python
import seaborn
%matplotlib inline
seaborn.countplot("Species",data=iris)
```

```python
seaborn.barplot(x='Species', y='PetalLengthCm', data=iris)
```

```python
seaborn.boxplot(x='Species', y='PetalLengthCm', data=iris)
```

```python
seaborn.distplot(iris['PetalLengthCm'])
```

```python
iris_vir = iris[iris.Species == 'Iris-virginica']
iris_s = iris[iris.Species == 'Iris-setosa']
iris_ver = iris[iris.Species == 'Iris-versicolor']
seaborn.distplot(iris_vir['PetalWidthCm'],label='vir').set(ylim=(0,12))
seaborn.distplot(iris_s['PetalWidthCm'],label='s')
seaborn.distplot(iris_ver['PetalWidthCm'],label='ver').legend()
```

```python
g = seaborn.FacetGrid(iris, hue='Species')
g.map(seaborn.distplot, 'PetalWidthCm').add_legend()
```

```python
g = seaborn.FacetGrid(iris, row='Species')
g.map(seaborn.distplot, 'PetalWidthCm').add_legend()
```

```python
g = seaborn.FacetGrid(iris, col='Species')
g.map(seaborn.distplot, 'PetalWidthCm').add_legend()
```

```python
seaborn.regplot(x='PetalWidthCm', y='PetalLengthCm', data=iris)
```

```python
g = seaborn.FacetGrid(iris, hue='Species')
g.set(xlim=(0,2.5))
g.map(seaborn.regplot, 'PetalWidthCm', 'PetalLengthCm').add_legend()
```

```python
import matplotlib.pyplot as plt
g = seaborn.FacetGrid(iris, hue='Species')
g.set(xlim=(0,2.5))
g.map(plt.scatter, 'PetalWidthCm', 'PetalLengthCm').add_legend()
```
